```python
Matplotlib is a Plotting for the python programming languaage and its numerical
mathematics extension Numpy.

some of the major pros of matplotlib are:
   . Generally easy to get started for simple plots
   . support for custom labels and texts
   . great control of very element in a figure
   . high-quality output in many formats
   . very customizable in general

   . plt.bar():make a bar plot.
   . plt.plot(): can be used to make a line graph.

   . plt.subplots(): Add a support to the current figure. it returns two arguments fig and ax
    where fig figure and ax can be either a single axes object or an array of axes objects.

   . plt.xticks(): A list of positions at which ticks should be placed.
   . plt.xlabel(label_name):Set the label for the x-axis.
   . plt.ylabel(label_name):Set the label for the y-axis.
   . plt.show():Display a figure.
   . plt.grid():configure the grid lines. 
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame({"Length": np.random.randn(500)})
df.head()
```

### Histogram

```python
df['Length'].hist()
plt.show()
```

```python
plt.hist(df['Length'])
plt.show()
```

```python
plt.hist(df['Length'])
plt.show()
```

```python
plt.hist(df['Length'], bins=20)
plt.show()
```

```python
plt.hist(df['Length'], bins=10)
plt.xlabel("Length.")
plt.ylabel("count")
plt.title("Histogram of length")
plt.show()
```

```python
#horizontal orienatio
plt.hist(df['Length'], bins=10, orientation='horizontal')
plt.xlabel("Length in mm.")
plt.ylabel("Count")
plt.title("Histogram of piece length")
plt.show()
```

### Pie Chart

```python
list=[20,61,14,15,18,22]
lbs = ['Laptop','Mobile','Toothpaste','Shampoo','cricket','class']
```

```python
plt.pie(list,labels=lbs)
plt.show()
```

```python
plt.pie(list,labels=lbs,autopct='%1.1f%%')
plt.show()
```

```python
plt.pie(list,labels=lbs,autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
plt.pie(list,labels=lbs,autopct='%1.1f%%',shadow=True)
plt.axis("equal")
plt.show()
```

```python
plt.pie(list,labels=lbs,autopct='%1.1f%%',shadow=True,radius=1.5)
plt.axis("equal")
plt.show()
```

```python
#counterclock and angle properties
plt.pie(list,labels=lbs,shadow=True,autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0,0,0])
plt.axis("equal")
plt.show()
```

```python
la = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215,130,205,210]
co = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0,)

plt.pie(sizes,explode=explode,labels=la,colors=co,autopct='%1.1f%%')

plt.axis('equal')
plt.show()
```

### Bar Plot

```python
marks_scored =[80,85,65,95,75]
roll_no= [11,12,13,14,15]
plt.bar(roll_no,marks_scored)
plt.xlabel('roll no')
plt.ylabel('marks scored')
plt.show()
```

```python
x = [2,6,10]
y = [11,15,8]

x2 = [3,9,12]
y2 = [6,15,9]
plt.bar(x, y)
plt.bar(x2, y2, color = 'g')

plt.title('bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
company=['google','amazon','tcs','orient']
revenue=[95,115,75,120]

plt.bar(company,revenue,label='revenue')
plt.ylabel("revenue(bln)")
plt.title('US technology stocks')
plt.legend()
plt.show()
```

```python
plt.barh(company,revenue,label="Revenue")
plt.ylabel("Revenue(bln)")
plt.title('US technology stocks')
plt.legend()
```

```python
company=['google','amazon','tesla','wipro']
revenue=[90,65,84,110]
profit=[25,42,65,23]

plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")

plt.title('US technology stocks')
plt.legend()
plt.show()
```

```python
company=['google','amazon','tesla','wipro']
revenue=[90,65,84,110]
profit=[25,42,65,23]

xpos = np.arange(len(company))
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(bln)")
plt.title('US technology stocks')
plt.legend()
plt.show()
```
