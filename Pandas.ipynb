```python
india_weather = pd.DataFrame({
"city": ["mumbai","delhi","banglore"],
"temperature": [32,40,35],
"humidity": [80,50,75]
})

india_weather
```

```python
us_weather = pd.DataFrame({
"city":["mumbai","london","bbsr"],
"temperature":[21,45,35],
"humidity":[65,54,25]
})
us_weather
```

```python
#concate two dataframes
df = pd.concat([india_weather, us_weather])
df
```

```python
#if you want continous index
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
df = pd.concat([india_weather, us_weather],axis=0)
df
```

```python
temperature_df = pd.DataFrame({
"city": ["mumbai","delhi","banglore",'hyderbad'],
"temperature":[32,45,52,61]})

temperature_df
```

```python
humidity_df = pd.DataFrame({
"city": ["delhi","mumbai","banglore"],
"humidity":[65,24,35]})
humidity_df
```

```python
# merge two dataframes with out explicity mention index
df = pd.merge(temperature_df, humidity_df, on='city')
df
```

```python
# merge two dataframes with out explictly mention index
df = pd.merge(temperature_df, humidity_df, on="city")
df
```

```python
# OUTER-JOIN
df = pd.merge(temperature_df, humidity_df,on='city',how='outer')
df
```
